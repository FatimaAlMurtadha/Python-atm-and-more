# Create variables to be used on the analysis

```python
# 1- Numerical columns
numerical_cols_gr = [
    "Age",
    "Hours_Studied",
    "Attendance",
    "Sleep_Hours",
    "Stress_Level",
    "Screen_Time",
    "Previous_GPA",
    "Tutoring_Sessions_Per_Week",
    "Exam_Anxiety_Score"
]
```

```python
# 2- Categorical columns
categorical_cols_gr = [
    "Gender",
    "Part_Time_Job",
    "Study_Method",
    "Diet_Quality",
    "Internet_Quality",
    "Extracurricular",
    "Family_Income_Level",
    "Grade"
]
```

```python
# 3- Subsets
df_num_gr = df_cleaned_gr[numerical_cols_gr]
df_cat_gr = df_cleaned_gr[categorical_cols_gr]
```

```python
# 4- Grade as ordered category
grade_order = ["A", "B", "C", "D", "Fail"]
df_cleaned_gr["Grade"] = pd.Categorical(df_cleaned_gr["Grade"], categories=grade_order, ordered=True)
```

```python
# 5- A numerical grade in order to make it easier to draw later
grade_map = {"A":5, "B":4, "C":3 , "D":2 , "Fail":1}
df_cat_gr["Grade_numeric"] = df_cat_gr["Grade"].map(grade_map)
```

# 14-05-2026 VISUALIZATION + DATA STORYTELLING
## 1. Countplot - Grade Distribution

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df_cleaned_gr, x="Grade", order=["A","B","C","D","Fail"])
plt.title("Distribution of Grades")
plt.show()
```

### DATA STORYTELLING - Correlation Heatmap ( Student Performance Grade) 
- More hours studied + Tutoring sessions per week -> higher Grade ( strong - positive - logic).
- More exam anxiety score + stress level -> less Grade ( strong - negative - logic)
- Better Previous GPA + Attendance -> good Grade as before.
- Screen time + sleep hours -> Grade ( weak - negative - logic)
- Age -> Grade ( NO correlation at all) 

## Boxplot (Grade vs Study_Method)

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df_cleaned_gr,
    x="Study_Method",
    y="Grade"
)
plt.title("Study Method vs Grade")
plt.xlabel("Study_Method")
plt.ylabel("Grade")
plt.show()
```

### DATA STORYTELLING - ( Student Performance Grade vs Study Method) 
1. All three study methods (Online, Offline, Hybrid) follow a very similar grade pattern:
- Most students achieve A, followed by B  then C  
- Very few students fall into D or Fail
`This indicates that study method is not the primary driver of academic performance in this dataset.`

2. Although the overall pattern is similar, there are subtle but meaningful differences:
Offline
- Slightly higher concentration of A grades  
- More stable performance (lower variance)  
- Fewer low-grade outliers  
Hybrid
- Middle-ground performance  
- Balanced distribution between A and B  
- Moderate variance  
Online
- Highest variability  
- More outliers in the lower grades (C, D, Fail)  
- Indicates that some students struggle more in online environments

- B = 4.3 hours  
- C = 3.2 hours  
- D = 2.1 hours  
- Fail = 1.5 hours
`The difference between A and Fail is dramatic — more than 4 hours per day.
This reinforces that effort is the strongest academic factor in this dataset.`
- Study hours are a direct and reliable indicator of academic success  
- Students who dedicate more time to studying consistently achieve higher grades  
- Students with low study hours are at higher risk of poor performance  
- This insight can guide interventions, tutoring programs, and personalized study plans  
- In the FinalScore dataset, Hours_Studied had the strongest positive correlation  
- In the Grade dataset, it shows the clearest separation between grade categories

# Correlation Matrix

```python

plt.figure(figsize=(10,8))
sns.heatmap(df_num_gr.corr(), annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix (Numerical Variables)")
plt.show()
```

### DATA STORYTELLING - ( Student Performance - Correlation matrix)
- Academic effort variables correlate positively with each other, while psychological stress variables correlate negatively with academic behaviors.
- Strong Positive Correlations

    - HoursStudied + TutoringSessionsPerWeek ( 0.27) - students who study more tend to seek more academic support.
    - Attendance + Previous_GPA ( 0.20) - students with strong academic history maintain higher attendance.
    - Hours_Studied + Attendance ( 0.18) - students who attend more also study more.
- Strong Negative Correlations

    - StressLevel + HoursStudied ( –0.29) - higher stress is associated with fewer study hours.
    - ExamAnxiety + HoursStudied ( –0.29) - students who study more feel more prepared and less anxious.
    - StressLevel + SleepHours ( –0.25) - less sleep is linked to higher stress - a classic psychological pattern.
    - ExamAnxiety + SleepHours ( –0.22) - poor sleep increases anxiety, which then affects performance.
- Weak or No Correlations
    - Age: shows almost no meaningful correlation with any variable. This confirms that age is not a performance factor in this dataset.
    - Screen_Time: screen time has a weak negative relationship with study hours and performance‑related variables.

# Pairplot ( Psychological and Academic factors)

```python

sns.pairplot(df_cleaned_gr[["Hours_Studied", "Stress_Level", "Exam_Anxiety_Score", "Previous_GPA"]])
plt.show()
```

### DATA STORYTELLING - Pairplot ( Psychological and Academic factors)
- Students who study more hours tend to: have lower stress levels, have lower exam anxiety and have higher previous GPA.
- High study effort -> low psychological strain -> strong academic history.
- Stress shows: a negative relationship with Hours_Studied, a positive relationship with Exam_Anxiety, and a negative relationship with Previous_GPA
- Stress reduces study consistency and is more common among students with weaker academic backgrounds.
- Exam anxiety behaves similarly to stress but with even clearer patterns:
    - Higher anxiety -> fewer study hours.
    - Higher anxiety -> higher stress.
    - Higher anxiety -> lower previous GPA.
- Anxiety increases stress -> stress reduces study hours -> reduced study hours lower performance -> performance issues increase anxiety.
- Previous GPA correlates positively with: Hours_Studied, Lower stress and Lower exam anxiety.
- Students who performed well in the past maintain strong habits and emotional stability.
- Academic strength = high study effort + low psychological strain.
- Academic risk = low study effort + high psychological strain.
