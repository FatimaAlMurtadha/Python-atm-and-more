## Boxplot (Grade vs Study_Method)

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    data=df_cleaned_gr,
    x="Study_Method",
    y="Grade"
)
plt.title("Study Method vs Grade")
plt.xlabel("Study_Method")
plt.ylabel("Grade")
plt.show()
```

### DATA STORYTELLING - ( Student Performance Grade vs Study Method) 
1. All three study methods (Online, Offline, Hybrid) follow a very similar grade pattern:
- Most students achieve A  
- Followed by B  
- Then C  
- Very few students fall into D or Fail
`This indicates that study method is not the primary driver of academic performance in this dataset.`

2. Although the overall pattern is similar, there are subtle but meaningful differences:
Offline
- Slightly higher concentration of A grades  
- More stable performance (lower variance)  
- Fewer low-grade outliers  
Hybrid
- Middle-ground performance  
- Balanced distribution between A and B  
- Moderate variance  
Online
- Highest variability  
- More outliers in the lower grades (C, D, Fail)  
- Indicates that some students struggle more in online environments
