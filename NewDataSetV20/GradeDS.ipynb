```python
df_cleaned = df.drop(columns=["Student_ID"])
df_cleaned.shape # (8000, 17)
```

# Create variables to be used on the analysis

```python
# 1- Numerical columns
numerical_cols2 = [
    "Age",
    "Hours_Studied",
    "Attendance",
    "Sleep_Hours",
    "Stress_Level",
    "Screen_Time",
    "Previous_GPA",
    "Tutoring_Sessions_Per_Week",
    "Exam_Anxiety_Score"
]
```

```python
# 2- Categorical columns
categorical_cols2 = [
    "Gender",
    "Part_Time_Job",
    "Study_Method",
    "Diet_Quality",
    "Internet_Quality",
    "Extracurricular",
    "Family_Income_Level",
    "Grade"
]
```

```python
# 3- Subsets
df_num = df_cleaned[numerical_cols2]
df_cat = df_cleaned[categorical_cols2]
```

```python
# 4- Grade as ordered category
grade_order = ["A", "B", "C", "D", "Fail"]
df_cleaned["Grade"] = pd.Categorical(df_cleaned["Grade"], categories=grade_order, ordered=True)
```

```python
# 5- A numerical grade in order to make it easier to draw later
grade_map = {"A":5, "B":4, "C":3 , "D":2 , "Fail":1}
df_cat["Grade_numeric"] = df_cat["Grade"].map(grade_map)
```
