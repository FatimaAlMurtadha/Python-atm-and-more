### agg() vs transform()

```python
df_cleaned["Final_Score"].agg([
    "mean", "median", "min","max","std"
])
```

```python
df_cleaned[
    [
        "Hours_Studied", "Stress_Level", "Final_Score"
    ]
].agg(["mean", "min", "max"])
```

```python
df_cleaned.groupby("Gender").agg({
    "Final_Score": [
        "mean", "std", "min","max"],
    "Hours_Studied": "mean",
    "Stress_Level":"mean"
})
```

```python
df_cleaned["Gender_Avg_Score"] = (
    df_cleaned.groupby("Gender")["Final_Score"].transform("mean")
)
df_cleaned
```

```python
df_cleaned["Score_Difference"] = (
    df_cleaned["Final_Score"] - df_cleaned["Gender_Avg_Score"]
)
df_cleaned
```

```python
df_cleaned["Score_Z"] = (
    (
        df_cleaned["Final_Score"]
        - df_cleaned["Final_Score"].mean()
    )
    /
    df_cleaned["Final_Score"].std()
)
df_cleaned
```

```python
df_cleaned["Method_Avg_Score"] = (
    df_cleaned.groupby("Study_Method")["Final_Score"]
    .transform("mean")
)
```
