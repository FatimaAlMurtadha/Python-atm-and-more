```python
# Grouping
# Aggregation 
df1 = pd.DataFrame({'name': ['fatima','sukaina','ahmed','sama','zaid']})
df1
# upper case
df1['upper_name'] = df1['name'].str.upper()
df1
# lower Case
df1['lower_name'] = df1['name'].str.lower()
df1
```

```python
# Contains
df2 = pd.DataFrame({'gmail': ['fatima@gmail.com','sukaina','ahmed@gmail.com','sama@gmail.com','zaid']})
df2['has_gmail'] = df2['gmail'].str.contains('@gmail.com')
df2
```

```python
# Melt 
df3 = pd.DataFrame({'students':['Fatima','Ahmed','Sukaina'],
                    'math':[90,95,100],
                    'science':[85,99,100]})
# change R to C - takes df, id_vars "the columns that we want them to have the same name as before"
# convert to4
df_melted = pd.melt(df3, 
                    id_vars='students',
                    value_vars=['math','science'],
                    var_name='subjects',
                    value_name='marks',
                    ignore_index=False)
df_melted
```

```python
# povet

df3_povet = df_melted.pivot(index='students', columns='subjects', values='marks')
df3_povet
```

```python
#
df_pivot_table = df_melted.pivot_table(index='students', columns='subjects', values='marks', aggfunc='mean')
df_pivot_table
```
