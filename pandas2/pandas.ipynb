```python
# File -> Data
df = pd.read_csv('./FINAL_DATASET.csv')
df
```

```python
# A column
print(df['category'])
```

```python
# More than 1 column
print(df[['category','gender']])
```

```python
# Row [n] iloc takes the index ignoring the naming
df.iloc[10]
```

```python
# index depending on the real index name  
# loc [ R. labelName]
df.loc[10]
```

```python
# iloc [R,C]
df.iloc[0,5]
```

```python
# All first n elements / ending not included
df.iloc[3:10]
```

```python
# Condition - filter df[df[condition]]
df[df['confidence_score'] > 0.89]
```

```python
# at == loc [R, labelName]
# iat == iloc [R, C ]
# at + iat faster 
df.iat[0,5]
```
